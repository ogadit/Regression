```python
from dataAnalysis import LinReg
import csv
```

### Importing Data from LinReg.csv and making arrays

```python
x = []
y = []

with open("LinReg.csv", 'r', newline='') as file:
    reader = csv.reader(file)
    next(reader) #Skip header
    for row in reader:
        x.append(float(row[0]))
        y.append(float(row[1]))

obj = LinReg(x, y)
```
